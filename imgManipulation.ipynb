```python
cb_image = cv.imread("./unnamed.jpeg", cv.IMREAD_COLOR)
cb_image = cb_image[:,:,::-1]
plt.imshow(cb_image)
```

```python
croped_Region = cb_image[100:300, 200:400]
plt.imshow(croped_Region)
```

```python
resized_croped_Region = cv.resize(croped_Region, None, fx=2, fy=2)
plt.imshow(resized_croped_Region)
```

```python
desired_width = 100
desired_height = 200
dim = (desired_width, desired_height)

resized_croped_Region = cv.resize(croped_Region, dsize=dim, interpolation = cv.INTER_AREA)
plt.imshow(resized_croped_Region)
```

```python
desired_width = 100
aspect_ratio = desired_width/croped_Region.shape[1]
desired_height = int(croped_Region.shape[0]* aspect_ratio)

dem = (desired_width, desired_height)

#resize image 
resized_croped_Region = cv.resize(croped_Region, dsize=dim, interpolation=cv.INTER_AREA)
plt.imshow(resized_croped_Region)
```

```python
#swap channel orders

resized_croped_Region_2x = resized_croped_Region[:,:,::-1]
cv.imwrite("resize_cropped_img.png", resized_croped_Region_2x)

Image(filename="./resize_cropped_img.png")
```

```python
#flipping Images
img=cv.imread("./unnamed.jpeg", cv.IMREAD_COLOR)
plt.imshow(img)
```

```python
img_bgr_hor = cv.flip(img,1)
img_bgr_ver = cv.flip(img,0)
img_bgr_both = cv.flip(img,-1)


#show images
plt.figure(figsize=[18,5])
plt.subplot(141);plt.imshow(img_bgr_hor);plt.title("Horizontal")
plt.subplot(142);plt.imshow(img_bgr_ver);plt.title("Vertical")
plt.subplot(143);plt.imshow(img_bgr_both);plt.title("Both Flipped")
plt.subplot(144);plt.imshow(img);plt.title("Original")
```
