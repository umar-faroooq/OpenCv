```python
from google.colab.patches import cv2_imshow
import cv2 as cv
```

```python
img = cv.imread("/content/babar.jpg")
```

```python
cv2_imshow(img)
```

```python
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
cv2_imshow(img_gray)
```

```python
img_invert = cv.bitwise_not(img_gray)
```

```python
cv2_imshow(img_invert)
```

```python
img_smoothing = cv.GaussianBlur(img_invert, (21, 21) , sigmaX=0, sigmaY=0)
```

```python
cv2_imshow(img_smoothing)
```

```python
def dodgeV2(x, y):
  return cv.divide(x, 255 - y, scale=256)
```

```python
final_image = dodgeV2(img_gray, img_smoothing)
type(final_image)
#
```

```python
cv2_imshow(final_image)
```
